## 이웃한 snp를 단순문자열로 결합하여 사용한 모델
> 예시 : SNP_01 + SNP_02

```python
import pandas as pd
import random
import os
import numpy as np

from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```python
raw_data = pd.read_csv('./dataset/train.csv')
data = raw_data.copy()
data.info()
```

```python
data = data.drop(columns=['id','father','mother','gender'])
data.head()
```

```python
data['SNP_16'] = data[['SNP_01', 'SNP_02']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_17'] = data[['SNP_02', 'SNP_03']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_18'] = data[['SNP_03', 'SNP_04']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_19'] = data[['SNP_04', 'SNP_05']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_20'] = data[['SNP_05', 'SNP_06']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_21'] = data[['SNP_06', 'SNP_07']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_22'] = data[['SNP_07', 'SNP_08']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_23'] = data[['SNP_08', 'SNP_09']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_24'] = data[['SNP_09', 'SNP_10']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_25'] = data[['SNP_10', 'SNP_11']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_26'] = data[['SNP_11', 'SNP_12']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_27'] = data[['SNP_12', 'SNP_13']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_28'] = data[['SNP_13', 'SNP_14']].apply(lambda x: ' '.join(x), axis=1)
data['SNP_29'] = data[['SNP_14', 'SNP_15']].apply(lambda x: ' '.join(x), axis=1)
```

```python
y_data = data['class']
x_data = data.drop(columns = ['class'])
x_test = pd.read_csv('./dataset/test.csv')
x_test = x_test.drop(columns = ['id','father','mother','gender'])
```

```python
x_test['SNP_16'] = x_test[['SNP_01', 'SNP_02']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_17'] = x_test[['SNP_02', 'SNP_03']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_18'] = x_test[['SNP_03', 'SNP_04']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_19'] = x_test[['SNP_04', 'SNP_05']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_20'] = x_test[['SNP_05', 'SNP_06']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_21'] = x_test[['SNP_06', 'SNP_07']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_22'] = x_test[['SNP_07', 'SNP_08']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_23'] = x_test[['SNP_08', 'SNP_09']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_24'] = x_test[['SNP_09', 'SNP_10']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_25'] = x_test[['SNP_10', 'SNP_11']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_26'] = x_test[['SNP_11', 'SNP_12']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_27'] = x_test[['SNP_12', 'SNP_13']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_28'] = x_test[['SNP_13', 'SNP_14']].apply(lambda x: ' '.join(x), axis=1)
x_test['SNP_29'] = x_test[['SNP_14', 'SNP_15']].apply(lambda x: ' '.join(x), axis=1)
```

```python
model = RandomForestClassifier(
            max_depth= 5, ## 개별 나무의 최대 깊이
            random_state=1234)
```

```python
class_le = preprocessing.LabelEncoder()
snp_le = preprocessing.LabelEncoder()
snp_col = ['SNP_01', 'SNP_02', 'SNP_03', 'SNP_04','SNP_05','SNP_06','SNP_07','SNP_08','SNP_09','SNP_10','SNP_11','SNP_12','SNP_13','SNP_14','SNP_15','SNP_16','SNP_17','SNP_18','SNP_19','SNP_20','SNP_21','SNP_22','SNP_23','SNP_24','SNP_25','SNP_26','SNP_27','SNP_28','SNP_29']
```

```python
snp_data = []
for col in snp_col:
    snp_data += list(x_data[col].values)
y_data = class_le.fit_transform(y_data)
snp_le.fit(snp_data)

for col in x_data.columns:
    if col in snp_col:
        x_data[col] = snp_le.transform(x_data[col])
        x_test[col] = snp_le.transform(x_test[col])
```

```python
model.fit(x_data,y_data)
preds = model.predict(x_test)
x_test
```

```python
submit = pd.read_csv('./dataset/sample_submission.csv')
submit['class'] = class_le.inverse_transform(preds)
submit.to_csv('./dataset/submit_1.csv', index=False)
```
